```python
# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt

print("Mari mulai dengan operasi dasar!")

# =================
# 1. Basic Arrays
# =================

# Membuat array sederhana
print("\n--- Basic Arrays ---")
data = np.array([1, 2, 3, 4, 5])
print("Array sederhana:", data)
print("Shape (bentuk) array:", data.shape)

# Operasi matematika dasar
print("\nOperasi matematika:")
print("Plus 2:", data + 2)
print("Kali 3:", data * 3)
print("Pangkat 2:", data ** 2)
```

```python
# =================
# 2. Time Series
# =================

print("\n--- Time Series ---")
# Membuat data waktu
waktu = np.linspace(0, 1, 1000)  # 1000 titik dari 0 sampai 1
print("Time array pertama:", waktu[:5], "...")  # tampilkan 5 nilai pertama

# Membuat gelombang sintetis sederhana
frekuensi = 30  # Hz
amplitudo = np.sin(2 * np.pi * frekuensi * waktu) * np.exp(-2 * waktu)
print("Amplitudo pertama:", amplitudo[:5], "...")
```

```python
# =================
# 3. Basic Plotting
# =================

print("\n--- Basic Plotting ---")
# Plot sederhana
plt.figure(figsize=(10, 6))
plt.plot(waktu, amplitudo, 'b-', lw=1)
plt.title('Synthetic Seismic Trace')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# =================
# 4. Simple Processing
# =================
from scipy import signal

print("\n--- Simple Processing ---")
# Smoothing data
window = 51  # window length harus ganjil
smoothed = signal.savgol_filter(amplitudo, window, 3)

# Plot original vs smoothed
plt.figure(figsize=(12, 6))
plt.plot(waktu, amplitudo, 'b-', label='Original', alpha=0.5)
plt.plot(waktu, smoothed, 'r-', label='Smoothed')
plt.title('Original vs Smoothed Data')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# =================
# 6. Advanced: Functions
# =================

print("\n--- Advanced: Functions ---")
# Contoh function sederhana
def buat_trace(n_samples=1000, freq=30):
    t = np.linspace(0, 1, n_samples)
    trace = np.sin(2 * np.pi * freq * t) * np.exp(-2 * t)
    return t, trace

# Gunakan function
t_new, trace_new = buat_trace(freq=20)
plt.figure(figsize=(10, 6))
plt.plot(t_new, trace_new)
plt.title('Trace from Function (freq=20Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
